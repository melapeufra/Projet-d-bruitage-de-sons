```python
#Question 6

N = f_s = 4096
#N=10
T = 1
l_1 = 400
l_2 = 600
l_3 = 1000

i = complex(0,1)

#Création des parties réelles des vecteurs Epsilon l (pour l = 400, 600, 1000) -  Re_(Epsl)
eps_400 = []
for n in range (N):
    c = np.exp((2*i*np.pi*n*400)/N)  # On applique la formule de la n_ième coordonée du vecteur Epsilon l 
    eps_400.append(c.real)           # Ici, on prend la partie réelle et on l'ajoute au vecteur 

eps_600 = []
for n in range (N):
    c = np.exp((2*i*np.pi*n*600)/N)  # On applique la formule de la n_ième coordonée du vecteur Epsilon l 
    eps_600.append(c.real)           # Ici, on prend la partie réelle et on l'ajoute au vecteur 
    
eps_1000 = []
for n in range (N):
    c = np.exp((2*i*np.pi*n*1000)/N)  # On applique la formule de la n_ième coordonée du vecteur Epsilon l 
    eps_1000.append(c.real)           # Ici, on prend la partie réelle et on l'ajoute au vecteur 

# On change les listes en vecteur :
eps_400 = np.array(eps_400)
eps_600 = np.array(eps_600)
eps_1000 = np.array(eps_1000)

# On transforme les parties réelles des vecteurs Epsilon l en son : 
wavfile.write('eps_400.wav', f_s,eps_400)
wavfile.write('eps_600.wav', f_s,eps_600)
wavfile.write('eps_1000.wav', f_s,eps_1000)

# Après les avoir écouté on remarque que plus l augmente plus le son devient aigu...

# On cherche la valeur de Re(Epsl)_chapeau 
eps_400_chap = npft.fft(eps_400)
eps_600_chap = npft.fft(eps_600)
eps_1000_chap = npft.fft(eps_1000)

print ("eps_400_chap = ",eps_400_chap, "eps_600_chap = ", eps_600_chap, "eps_1000_chap = ", eps_1000_chap)
```

```python
def complexe_modulo(z):
     a = z.real
     b = z.imag
     return math.sqrt(a**2+b**2)
```

```python
#Question 9 

u0_chap = npft.fft(u0)
u1_chap = npft.fft(u1)

modu_u0 = []
for k in range (len(u0_chap)) :
    modu_u0.append(complexe_modulo(u0_chap[k]))

modu_u1 = []
for k in range (len(u1_chap)) :
    modu_u1.append(complexe_modulo(u1_chap[k]))

plt.plot(n_u0,modu_u0)
plt.show()

plt.plot(n_u1,modu_u1)
plt.show()

#Graphiquement, on voit que seulement sur le graph du module de u0_chapeau, il y a des valeurs extrèmement proche
# de 0. Donc seul u0 (et non u1) appartient à U_k0 - On peut donc essayer de réparer u0 

# On cherche M = k0 pour la fonction filtrage, le moment 
# à partir duquel les valeurs sont très proches de 0 (<10**(-15)) : 

def cherche_k0 (mod) :
    k = 0
    for n in mod :
        if n < 10**(-15) :
            break
        else :
            k +=1
    return k

print(cherche_k0(modu_u0))

#On applique la fonction du poly pour restaurer le son :

def filtrage (M,y):
    ychap = np.fft.fft(y)
    N = ychap.size
    mask = np.ones(N)
    coord = np.arange(M+1,N-M+1)
    mask[coord] = 0
    yfiltchap = mask * ychap
    yfilt = npft.ifft(yfiltchap)
    yfilt = yfilt.real
    return yfilt

M = cherche_k0(modu_u0)
yfilt = filtrage(M, y0)  # Vecteur restauré

plt.plot(n_u0,yfilt)
plt.show()


#f_s_u0 = f_s_y0

# On transforme le vecteur restauré en son :

wavfile.write('yfilt.wav', f_s_u0, yfilt)
```
